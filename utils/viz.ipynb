```python
import json
import matplotlib.pyplot as plt
import numpy as np

# Load the data
with open('../data/affect.json', 'r') as f:
  data = json.load(f)

# Extract the first element from each list
first_values = []
for key, values in data.items():
  if values and values[0] is not None:  # Skip empty lists and None values
    first_values.append(values[0])

# Create the plot
plt.figure(figsize=(10, 6))
plt.hist(first_values, bins=20, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Distribution of First Values from Affect Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True, alpha=0.3)
plt.show()
```
